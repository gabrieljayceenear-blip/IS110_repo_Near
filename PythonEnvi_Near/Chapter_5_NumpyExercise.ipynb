

EXERCISE A



```python
import numpy as np

# creating an array list with values
nestedArrayList = [[10, 20, 30], [40, 50, 60]]

# converting the list into a numpy ndarry
A = np.array(nestedArrayList)

# using property to get the shape of the list (rows, columns)
print(A.shape)

# using property to get the maximum number inside the array list
print(A.max())

# passing the object into the numpy function to get the same result
print(np.max(A))
```

EXERCISE B

```python
B = np.ones((5, 5), dtype=int) # creating a 5x5 shape with data type of 'int' to remove devimals

print(B.shape)  # displaying the shape of the array created above
print(B.size)   # shows the total size of the array 5x5 = 25
print(B.ndim)   # checking the dimension of the array
```

```python
# creating an array from -5 to 6 because in array the function stops right before it reaches the stated value
C = np.arange(-5, 6)

# multiplying each values of the array from -5 to 5 by 2
result = C * 2

# displaying the result with a value of previous calculation
print(result)
```
