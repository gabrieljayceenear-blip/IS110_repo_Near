"==" checks if the values are similar "is" check if the memory location and values are similar

Challenge 1 (Answer)

```python
total = 0
count = 0

while True:
    num_input = input("Enter a number: ")
    if num_input == "done":
        break
    try:
        num = float(num_input)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")

if count > 0:
    average = total / count
    print(total, count, average)
else:
    print("No numbers were entered.")
```

Challenge 2 (Answer)

```python
largest = None
smallest = None

while True:
    num_input = input("Enter a number: ")
    if num_input == "done":
        break
    try:
        num = float(num_input)
        if largest is None or num > largest:
            largest = num
        if smallest is None or num < smallest:
            smallest = num
    except ValueError:
        print("Invalid input")

if largest is not None and smallest is not None:
    print("Maximum:", largest)
    print("Minimum:", smallest)
else:
    print("No numbers were entered.")
```
