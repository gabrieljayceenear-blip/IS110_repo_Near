Exercise 1 (Answer)

```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
hours1 = 40

def compute_pay(hours, rate):
  if hours > 40:
    normal_pay = float(hours1)*float(rate)
    overtime_rate = (float(rate)*1.5)*float(hours - 40)
    pay =  float(normal_pay) + float(overtime_rate)
    print("Pay:", float(pay))
  if hours < 41:
    pay =  float(rate)*float(hours)
    print("Pay:", float(pay))

compute_pay(hours, rate)
```

```python
score = int (input("Enter Score: "))

def computegrade(score):
  if score >= 97 and score <= 100:
    print("Grade: 1.0")
  elif score >= 93 and score <= 97:
    print("Grade: 1.25")
  elif score >= 89 and score <= 93:
    print("Grade: 1.5")
  elif score >= 85 and score <= 89:
    print("Grade: 1.75")
  elif score >= 81 and score <= 85:
    print("Grade: 2.0")
  elif score >= 77 and score <= 81:
    print("Grade: 2.25")
  elif score >= 74 and score <= 77:
    print("Grade: 2.5")
  else:
    print("Grade: 3.0")

computegrade(score)
```
