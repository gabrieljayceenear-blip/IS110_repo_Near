```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
hours1 = 40

if hours > 40:
     normal_pay = float(hours1)*float(rate)
     overtime_rate = (float(rate)*1.5)*float(hours - 40)
     pay =  float(normal_pay) + float(overtime_rate)
     print("Pay:", float(pay))
if hours < 41:
    pay =  float(rate)*float(hours)
    print("Pay:", float(pay))
```
